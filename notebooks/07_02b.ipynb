Difflib digunakan untuk menemukan perbedaan antara dua atau lebih urutan teks, yang bisa berupa baris teks dalam file, string, atau daftar kata. Dengan menggunakan difflib, kita dapat dengan mudah melihat bagaimana dua teks itu berbeda atau sama, yang sangat berguna untuk berbagai aplikasi seperti deteksi plagiarisme, menggabungkan perubahan dalam pengembangan perangkat lunak, dan banyak lagi.

```python
import difflib
```

```python
fline = 'Hello. How are you? I am fine.'
gline = 'How are you, Lilian? I am doing well.'
```

```python
# Create a differ object (d)
d = difflib.Differ()
# To compare the sequences, we need to call the compare method off of the d object
diff = d.compare(fline, gline)
#Iterate over the sequence and print out the differences
for line in diff:
  print (line)
```
