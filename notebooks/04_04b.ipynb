### Defining plot color

```python
x = range(1,10)
y = [1,2,3,4,0.5,4,3,2,1]
```

```python
plt.bar(x,y)
```

First, create a list to define the width of the bars

```python
wide = [0.5,0.5,0.5,0.9,0.9,0.9,0.5,0.5,0.5]
color = ['salmon']

plt.bar(x,y, width=wide, color=color, align='center')
```

```python
z = [1,2,3,4,0.5]
plt.pie(z, autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

```python
color_theme1 = ['#7959E2', '#59A7E2', '#59E290', '#A2E259', '#E29559']
plt.pie(z, colors=color_theme1, autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

### Customizing line styles

```python
x1 = range(1,10)
y1 = [9,8,7,6,5,4,3,2,1]
plt.plot(x,y)
plt.plot(x1,y1)
```

```python
plt.plot(x,y, ls= 'solid', lw = 4)
plt.plot(x1,y1, ls= '--', lw=9)
```

### Setting plot markers

```python
plt.plot(x,y, marker='1', mew=20)
plt.plot(x1,y1, marker='+', mew=15)
```
