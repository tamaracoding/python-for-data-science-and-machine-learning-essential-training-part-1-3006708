```python
import numpy as np
from numpy.random import randn
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from matplotlib import rcParams
```

### Defining axes, ticks, and grids

```python
%matplotlib inline
rcParams['figure.figsize']=5,4
```

```python
x = range(1,11)
y = [1,2,3,4,5,0,4,3,2,1]

fig = plt.figure() #create blank figure
ax= fig.add_axes([.1,.1,1,1]) #add axes ([left, bottom, width, height])

ax.plot(x,y)
plt.show()
```

```python
x = range(1,10)
y = [1,2,3,4,0,4,3,2,1]

fig = plt.figure() #create blank figure
ax= fig.add_axes([.1,.1,1,1]) #add axes ([left, bottom, width, height])

ax.set_xlim([1,9]) #setting limit to axes 
ax.set_ylim ([0,5])

ax.set_xticks ([0,1,2,3,4,5,6,8,9,10]) #setting some tick marks
ax.set_yticks ([0,1,2,3,4,5])
ax.plot(x,y)
plt.show()
```

```python
fig = plt.figure() #create blank figure
ax= fig.add_axes([.1,.1,1,1]) #add axes ([left, bottom, width, height])

ax.set_xlim([1,9]) #setting limit to axes 
ax.set_ylim ([0,5])

ax.grid() #add a grid to the plot
ax.plot(x,y) 
plt.show()
```

### Generating multiple plots in one figure with subplots 

```python
fig = plt.figure()#create blank figure
fig,(ax1, ax2, ax3) = plt.subplots(1,3) #adjust the figure to include two subplots (1 row, 2 columns)
ax1.plot(x)
ax2.plot(x,y)
ax3.plot(y)
```

we have created one figure with two axes, and then we have generated plots within that one figure

```python
# Definisikan karakteristik makanan sebagai dictionary
makanan = {
    'enak': True,
    'murah': True,
    'sehat': True
}

# Memeriksa kondisi makanan
if makanan['enak'] and makanan['murah']:
    makanan['sehat'] = False
elif makanan['enak'] and makanan['sehat']:
    makanan['murah'] = False
elif makanan['murah'] and makanan['sehat']:
    makanan['enak'] = False

# Cetak hasil
print("Makanan enak:", makanan['enak'])
print("Makanan murah:", makanan['murah'])
print("Makanan sehat:", makanan['sehat'])
```
