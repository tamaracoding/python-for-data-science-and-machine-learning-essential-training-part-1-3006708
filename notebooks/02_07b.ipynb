### Grouping data by column index

```python
address = '/workspaces/python-for-data-science-and-machine-learning-essential-training-part-1-3006708/data/mtcars.csv'
```

```python
cars = pd.read_csv(address)
cars.head()
```

```python
cars = cars.rename(columns={'Unnamed: 0': 'cars_name'})
```

```python
cars.head()
```

How to group this dataframe by a particular column using groupby method

```python
cars_groups = cars.groupby(cars['cyl'])
```

So, say we want to generate a mean of the values, the numeric values in the dataframe as grouped by the cylinders column.

We need to pass in a parameter that says numeric only equal to true, because we only want to be looking at the numeric values since we're calculating a mean

```python
cars_groups.mean(numeric_only=True)
```

Say you want to know the correlation between the amount of gears with the fuel efficiency

```python
def efficientcar (s):
  return s.iloc[0]
```

```python
cars.sort_values('mpg', ascending=True).groupby(['gear','cyl']).agg(efficientcar)
```

Filter the data to find only 6 cilinder Maserati Bora

```python
six_cyl_cars=cars.loc[cars['cyl']==6,:]
six_cyl_cars
```

```python
six_cyl_cars.loc[six_cyl_cars['cars_name']=='Maserati Bora',:]
```

Cari mobil dengan horsepower terbesar dari masing-masing silinder

```python
def highestHPcar (s):
  return s.iloc[0]
```

```python
cars.sort_values('hp',ascending=False).groupby(['cyl']).agg(highestHPcar)
```

Bandingkan rata-rata hp mobil transmisi manual dengan transmisi otomatis

```python
manual_cars = cars.loc[cars['am']==0,:]
auto_cars = cars.loc[cars['am']==1,:]
```

```python
manual_mean = manual_cars['hp'].mean()
auto_mean = auto_cars['hp'].mean()
```

```python
print("Rata-rata Hp pada mobil manual:", manual_mean, "dan Rata-rata Hp pada mobil Auto:", auto_mean)
```
