## Creating labels and annotations

**Annotation (penjelasan)** adalah teks atau elemen tambahan yang ditempatkan pada grafik atau visualisasi data untuk memberikan konteks atau informasi tambahan. Anotasi biasanya digunakan untuk menjelaskan atau menyoroti bagian tertentu dari grafik, memberikan interpretasi, atau menyampaikan pesan yang lebih mendalam tentang data.

Tujuan: Memberikan penjelasan tambahan, konteks, atau interpretasi pada data.

Contoh:
Menambahkan catatan yang menjelaskan lonjakan data pada titik tertentu (misalnya, "Taman Hiburan XYZ dibuka pada 15 Juni").
Menyertakan penunjuk (pointer) yang menghubungkan teks dengan elemen spesifik pada grafik.
Menyertakan informasi tambahan seperti sumber data atau asumsi yang digunakan dalam analisis.

**Labelling (pelabelan)** adalah teks yang digunakan untuk mengidentifikasi elemen-elemen pada grafik atau dataset. Label biasanya mencakup nama sumbu, judul grafik, atau label data individu dalam grafik, seperti titik atau batang.

Tujuan: Mengidentifikasi elemen-elemen pada grafik atau dataset, sehingga memudahkan pembaca untuk memahami apa yang direpresentasikan.

Contoh:
Memberikan judul pada grafik (misalnya, "Jumlah Wisatawan di Florida Tengah per Bulan").
Menyertakan label pada sumbu x dan y (misalnya, sumbu x: "Bulan", sumbu y: "Jumlah Wisatawan").
Menyertakan label pada elemen data individu (misalnya, label pada batang dalam grafik batang yang menunjukkan nilai numerik).

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from pylab import rcParams
import seaborn as sb
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 12,4
sb.set_style('whitegrid')
```

### Labeling plot features
#### The functional method

```python
x = range(1,10)
y = [1,2,3,4,0.5,4,3,2,1]

plt.bar(x,y, width= [2,1,2,1,2,1,2,1,2])
plt.xlabel ('your x-axis label')
plt.ylabel ('your y-axis label')
```

```python
z = [1,2,3,4,0.5]
veh_type = ['bicycle','motorbike', 'car', 'van', 'stroller']
color_theme = ['#7959E2', '#59A7E2', '#59E290', '#A2E259', '#E29559']
plt.pie(z, labels= veh_type, colors=color_theme, autopct='%1.1f%%')
plt.title('vehicle distribution')
plt.show()
```

#### The object-oriented method

```python
address = '/workspaces/python-for-data-science-and-machine-learning-essential-training-part-1-3006708/data/mtcars.csv'
cars = pd.read_csv(address)
cars.head()
```

```python
cars = cars.rename(columns={'Unnamed: 0': 'car_names'})
cars.head()
```

```python
mpg = cars.mpg

fig = plt.figure() #create blank figure
ax = fig.add_axes ([.1,.1,1,1]) #add axes ([left, bottom, width, height])

mpg.plot(marker='o')

ax.set_xticks (range(32)) 
# 32 tick marks di sepanjang sumbu x
# Alternative: ax.set_xticks(range(len(cars)))
ax.set_xticklabels (cars.car_names, rotation = 60, fontsize = 'medium')
# We've got a little rotation in terms of the labels on the x-axis
ax.set_title('Miles per Gallon of Cars')
ax.set_xlabel ('Car Names')
ax.set_ylabel ('Miles per Gallon')
```

Some of the text of xlabel is piling on top of themselves, because we set the parameters with the rcParams at the top.

So, I chage rcParams['figure.figsize'] = 8,4 menjadi --> rcParams['figure.figsize'] = 12,4

### Adding a legend to your plot
#### The functional method

```python
plt.pie(z, colors=color_theme, autopct='%1.1f%%')
plt.legend(veh_type, loc='best')
plt.title('Distribution of Vehicle Type')
plt.show()
```

#### The object-oriented method

```python
fig = plt.figure()
ax = fig.add_axes ([.1,.1,1,1])

mpg.plot(marker ='o')

ax.set_xticks (range(32))

ax.set_xticklabels (cars.car_names, rotation = 60, fontsize = 'medium')
#We've got a little rotation in terms of the labels on the x-axis
ax.set_title('Miles per Gallon of Cars')
ax.set_xlabel ('Car Names')
ax.set_ylabel ('Miles per Gallon')

ax.legend(loc ='best')
```

```python
fig = plt.figure()
ax = fig.add_axes ([.1,.1,1,1])

mpg.plot(marker ='o')

ax.set_xticks (range(32))

ax.set_xticklabels (cars.car_names, rotation = 60, fontsize = 'medium')
#We've got a little rotation in terms of the labels on the x-axis
ax.set_title('Miles per Gallon of Cars')
ax.set_xlabel ('Car Names')
ax.set_ylabel ('Miles per Gallon')

ax.set_ylim ([0,45])

ax.annotate('Toyota Corolla', xy=(19, 33.9), xytext=(21,35), arrowprops=dict(facecolor='black', shrink = 0.5))

ax.legend(loc ='best')
```
