## Visualizing time series

```python
import numpy as np
from numpy.random import randn
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from pylab import rcParams
import seaborn as sns
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 5, 4
sns.set_style('whitegrid')
```

### The simplest time series plot

```python
address = '/workspaces/python-for-data-science-and-machine-learning-essential-training-part-1-3006708/data/Superstore-Sales.csv'
sales = pd.read_csv(address, index_col= 'Order Date',encoding='cp1252', parse_dates=True)
sales.head()
```

```python
sales['Order Quantity'].plot()
```

We can't make any sense of this, right? So what I'm going to do is I'm just going to create a small sample of this data so that we can plot something out that we're able to actually decipher

```python
sales_sample = sales.sample(n=100, random_state=25, axis=0)

plt.xlabel = ('Order Date')
plt.ylabel = ('Order Qty')

plt.title = ('Super Store Sale')

sales_sample['Order Quantity'].plot()
```
